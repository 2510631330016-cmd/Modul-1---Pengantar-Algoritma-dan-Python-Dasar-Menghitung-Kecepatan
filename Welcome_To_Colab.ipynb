```
import matplotlib.pyplot as plt

# PROGRAM SIMULASI GERAK JATUH BEBA
# Konsep Fisika: GLBB dengan percepatan gravitasi g = 9.8 m/s^2
# Rumus posisi: y = 1/2 * g * t^2

g = 9.8  # percepatan gravitasi (m/s^2)
waktu_total = int(input("Masukkan lama waktu jatuh (detik): "))

# List untuk menyimpan data waktu dan posisi
waktu = []
posisi = []

# Loop per detik
for t in range(waktu_total + 1):
    y = 0.5 * g * t**2   # rumus fisika GLBB

    # If condition: cek apakah benda sudah melewati 100 meter
    if y > 100:
        print(f"Pada detik ke-{t}, posisi sudah melewati 100 meter!")

    waktu.append(t)
    posisi.append(y)

# Tampilkan data hasil simulasi
print("\nHasil Simulasi:")
for t, y in zip(waktu, posisi):
    print(f"t = {t} s  -->  y = {y:.2f} m")

# Plotting grafik
plt.plot(waktu, posisi)
plt.title("Simulasi Gerak Jatuh Bebas")
plt.xlabel("Waktu (detik)")
plt.ylabel("Posisi (meter)")
plt.grid(True)
plt.show()
```
